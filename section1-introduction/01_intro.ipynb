```python
from pydantic import BaseModel
```

```python
class Person(BaseModel):
    first_name: str
    last_name: str
    age: int
```

```python
p = Person(first_name="Eisha", last_name="Enan", age=22)
```

```python
from pydantic import ValidationError
```

```python
try:
    Person(last_name="Enan")
except ValidationError as ex:
    print(ex)
```

```python
class Person(BaseModel):
    first_name: str
    last_name: str
    age: int

    @property
    def display_name(self):
        return f"{self.first_name} {self.last_name}"
```

```python
p = Person(first_name="Eisha", last_name="Enan", age=22)
```

```python
p.display_name
```

```python
p.display_name()
```

```python
try:
    Person(first_name="Eisha", last_name="Enan", age="not_a_Num")
except ValidationError as er:
    print(er)
```
