```python
from pydantic import BaseModel, ValidationError
```

```python
class Circle(BaseModel):
    center_x: int = 0
    center_y: int = 0
    radius: int = 0
    name: str | None = None
```

```python
Circle.model_fields
```

```python
c1 = Circle(radius=2)
c2 = Circle(name="unit circle")
```

```python
c1
```

```python
c2
```

```python
c1.model_dump()
```
