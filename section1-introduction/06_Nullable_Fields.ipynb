```python
from pydantic import BaseModel, ValidationError
```

```python
class Model(BaseModel):
    field: int
```

```python
try:
    Model(field=None)
except ValidationError as ex:
    print(ex)
```

```python
try:
    Model()
except ValidationError as ex:
    print(ex)
```

```python
class Model(BaseModel):
    field: int | None
```

```python
Model(field=None)
```

```python
class Model(BaseModel):
    field: int | None = None
```

```python
Model()
```

```python
from typing import Union

class Model(BaseModel):
    field: Union[int, None] = None # Another way of doing
```

```python
from typing import Optional
```

```python
class Model(BaseModel):
    field: Optional[int] # it is an int or None, but does not mean if it is not required
```

```python
class Model(BaseModel):
    field: Optional[int] = None
```
