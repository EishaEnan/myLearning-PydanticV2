```python
from pydantic import BaseModel, ValidationError
```

```python
class Coordinate(BaseModel):
    x: float
    y: float
```

```python
p1 = Coordinate(x=1.1, y=2.3)
p1
```

```python
Coordinate.model_fields
```

```python
type(p1.x)
```

```python
# what if the data we provide does not match the defined - pydantic tries to match
# It's called type coercion
```

```python
p2 = Coordinate(x=0, y="1.2")
p2
```

```python
type(p2.x)
```

```python
type(p2.y)
```

```python
class Model(BaseModel):
    field: str
```

```python
Model(field="Python")
```

```python
try:
    Model(field=100)
except ValidationError as ex:
    print(ex)
```

```python
class Contact(BaseModel):
    email: str
```
