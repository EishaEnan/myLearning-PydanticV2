```python
from dataclasses import dataclass
```

```python
try:
    @dataclass
    class Model:
        elements: list[int] = []
except ValueError as ex:
    print(ex)
```

```python
class log(BaseModel):
    dt: datetime = datetime.now(UTC)
    msg: str
```

```python
log(msg="Msg 1")
```

```python
log(msg="Msg 2")
```

```python
class log(BaseModel):
    dt: datetime = Field(default_factory=lambda: datetime.now(UTC))
    msg: str
```

```python
log(msg="Msg 1")
```
