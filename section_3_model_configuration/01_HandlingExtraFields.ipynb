```python
class Model(BaseModel):
    field_1: int
    
```

```python
m = Model(field_1=10, field_2=20)
m
```

```python
class Model(BaseModel):
    model_config = ConfigDict(extra="ignore")
    field_1: int
```

```python
m = Model(field_1=10, field_2=20)
m
```

```python
m.__dict__
```

```python
m.model_dump()
```

```python
m.model_extra
```

```python
class Model(BaseModel):
    model_config = ConfigDict(extra="forbid")
    field_1: int
```

```python
try:
    Model(field_1=10, field_2=20)
except ValidationError as ex:
    print(ex)
```
