```python
from enum import Enum
```

```python
class Color(Enum):
    red = "Red"
    green = "Green"
    blue = "Blue"
    
```

```python
Color.red
```

```python
type(Color.red)
```

```python
Color.red.value
```

```python
from pydantic import BaseModel, ConfigDict, ValidationError
```

```python
class Model(BaseModel):
    color: Color
```

```python
Model(color=Color.red)
```

```python
data = """
{
    "color": "Red"
}
"""
Model.model_validate_json(data)
```

```python
data = """
{
    "color": "Magenta"
}
"""
try:
    Model.model_validate_json(data)
except ValidationError as ex:
    print(ex)
```

```python
data = """
{
    "color": "Red"
}
"""
m = Model.model_validate_json(data)
```

```python
m.model_dump()
```

```python
m.model_dump_json()
```

```python
output = m.model_dump()
```

```python
output
```

```python
import json
try:
    json.dumps(output)
except TypeError as ex:
    print(ex)
```
