```python
from pydantic import BaseModel, ConfigDict, ValidationError
```

```python
class Model(BaseModel):
    model_config = ConfigDict(str_strip_whitespace=True)
    
    field: str
```

```python
m1 = Model(field=" Python ")
m1
```

```python
Model(field=" Python \t \n")
```

```python
class Model(BaseModel):
    model_config = ConfigDict(str_strip_whitespace=True, str_to_lower=True) # similar str_to_upper
    
    field: str
```

```python
Model(field=" Python \n")
```
