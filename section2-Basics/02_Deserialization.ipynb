```python
from pydantic import BaseModel, ValidationError
```

```python
class Person(BaseModel):
    first_name: str
    last_name: str
    age: int
```

```python
p = Person(first_name="Eisha", last_name="Enan", age=22)
```

```python
data = {
    'first_name': "Nykter",
    'last_name': "Stein",
    'age': 23
}

Person(**data)
```

```python
p = Person.model_validate(data)
p
```

```python
missing_data = {"last_name": "Ahdle"}
try:
    Person.model_validate(missing_data)
except ValidationError as ex:
    print(ex)
```

```python
# JSON - nums, str, list, obj, bool
data_json = '''
{
    "first_name": "Eisha",
    "last_name": "Enan",
    "age": 22
}
'''
```

```python
p = Person.model_validate_json(data_json)
p
```

```python
# JSON - nums, str, list, obj, bool
missing_json = '''
{
    "first_name": "Eisha",
}
'''

try:
    Person.model_validate(missing_json)
except ValidationError as ex:
    print(ex)

```
