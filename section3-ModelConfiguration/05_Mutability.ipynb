```python
from pydantic import BaseModel, ConfigDict, ValidationError
```

```python
class Model(BaseModel):
    field: int
```

```python
m = Model(field=10)
m
```

```python
m.field = 20
m
```

```python
class Model(BaseModel):
    model_config = ConfigDict(frozen=True)
    field: int
```

```python
m = Model(field=10)
```

```python
try:
    m.field = 20
except ValidationError as ex:
    print(ex)
```

```python
class Model(BaseModel):
    model_config = ConfigDict(frozen=False)
    field: int
```

```python
m = Model(field=10)
```

```python
try:
    d = {m: "trying to create a dictionary with Model object"}
except TypeError as ex:
    print(ex)
```
