```python
from pydantic import BaseModel, ConfigDict, ValidationError
```

```python
class Model(BaseModel):
    field: str
```

```python
try:
    Model(field=100)
except ValidationError as ex:
    print(ex)
```

```python
class Model(BaseModel):
    model_config = ConfigDict(coerce_numbers_to_str = True)
    field: str
```

```python
Model(field=1.5)
```
