```python
from pydantic import BaseModel, ValidationError
```

```python
class Person(BaseModel):
    first_name: str
    last_name: str
    age: int
```

```python
eisha = Person(first_name="Eisha", last_name="Enan", age=22)
nykter = Person(first_name="Nykter", last_name="Stein", age=23)
```

```python
eisha.__dict__
```

```python
eisha.model_dump()
```

```python
nykter.model_dump_json()
```

```python
nykter.model_json_schema()
```

```python
nykter.model_dump_json(indent=2)
```

```python
print(eisha.model_dump_json(indent=3))
```
