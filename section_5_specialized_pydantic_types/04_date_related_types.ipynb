```python
from pydantic import (
    BaseModel,
    ConfigDict,
    PastDatetime,
    PastDate,
    AwareDatetime,
    NaiveDatetime,
    ValidationError
)
from datetime import datetime
```

```python
class Model(BaseModel):
    dt: PastDatetime
```

```python
datetime.now()
```

```python
Model(dt=datetime(2025, 12, 16, 8, 35))
```

```python
class Model(BaseModel):
    dt: NaiveDatetime
```

```python
import pytz
from datetime import timedelta

local_one_hour_ago = datetime.now() - timedelta(hours=1)
local_one_hour_ago
```

```python
utc_one_hour_ago = local_one_hour_ago.astimezone(pytz.utc)
utc_one_hour_ago
```

```python
Model(dt=local_one_hour_ago)
```

```python
try:
    Model(dt=utc_one_hour_ago)
except ValidationError as ex:
    print(ex)
```

```python
class Model(BaseModel):
    dt: AwareDatetime
```

```python
Model(dt=utc_one_hour_ago)
```

```python
try:
    Model(dt=local_one_hour_ago)
except ValidationError as ex:
    print(ex)
```
