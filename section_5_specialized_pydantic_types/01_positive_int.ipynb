```python
from pydantic import BaseModel, PositiveInt, ValidationError
```

```python
class Circle(BaseModel):
    center: tuple[int, int] = (0, 0)
    radius: PositiveInt = 1


```

```python
c = Circle()
c
```

```python
try: 
    Circle(center=(0.5, 0.5), radius=0)
except ValidationError as ex:
    print(ex)
```
