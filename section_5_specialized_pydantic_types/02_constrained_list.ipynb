```python
from pydantic import BaseModel, PositiveInt, conlist, ValidationError
```

```python
class Sphere(BaseModel):
    center: conlist(int, min_length=2, max_length=3) = [0, 0]
    radius: PositiveInt = 1
```

```python
Sphere(center=(1, 2))
```

```python
Sphere(center=(1, 2, 3))
```

```python
try:
    Sphere(center=(1, 2, 3, 4))
except ValidationError as ex:
    print(ex)
```
