```python
from pydantic import BaseModel, Field, UUID4, ValidationError
```

```python
class Person(BaseModel):
    id: UUID4
```

```python
p = Person(id="f36f6941-56e1-4d69-a693-5261d0630c6b")
p
```

```python
p.model_dump()
```

```python
class Person(BaseModel):
    id: UUID4 = uuid4()
```

```python
p1 = Person()
p1
```

```python
p2 = Person()
p2
```

```python
class Person(BaseModel):
    id_: UUID4 = Field(alias="id", default=uuid4())
```

```python
p1 = Person()
p1
```

```python
p2 = Person()
p2
```

```python
class Person(BaseModel):
    id_: UUID4 = Field(alias="id", default_factory=uuid4)
```

```python
p1 = Person()
p1
```

```python
p2 = Person()
p2
```
